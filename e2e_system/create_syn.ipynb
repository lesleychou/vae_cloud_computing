```python
# Standard imports
import numpy as np
import pandas as pd
```

```python
def create_df(n_cont=2, n_cat=2, num_samples=10**6):
    # Create a dictionary with your columns
    data = dict()
    cont_names = []
    cat_names = []
    for i in range(n_cat):
        name = f'categorical_{i}'
        cat_names.append(name)
        data[name] = np.random.normal(loc=105, scale=2.5, size=num_samples).astype(int)
    for i in range(n_cont):
        name = f'continuous_{i}'
        cont_names.append(name)
        data[name] = np.random.normal(loc=550000, scale=250000, size=num_samples).astype(int)
    # data = {  
    #     'categorical_1': np.random.normal(loc=105, scale=2.5, size=10**6).astype(int),  # loc is the mean and scale is the standard deviation.  
    #     'categorical_2': np.random.normal(loc=105, scale=2.5, size=10**6).astype(int),  # loc is the mean and scale is the standard deviation.  
    #     'continuous_1': np.random.normal(loc=550000, scale=250000, size=10**6).astype(int),  
    #     'continuous_2': np.random.normal(loc=550000, scale=250000, size=10**6).astype(int)  
    # }  
      
    # Make sure the values are in the range you want
    for col in cat_names:
        data[col] = np.clip(data[col], 100, 109)
    for col in cont_names:
        data[col] = np.clip(data[col], 100000, 1000000)  
      
    # Create the dataframe  
    df = pd.DataFrame(data)
    
    return df
```
